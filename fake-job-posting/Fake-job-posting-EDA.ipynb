- required_education: Make it yes or no => yes for given and no for NaN<br/>

- function: Replace NaN with NG (Not given)<br/>

```python
# Replace all NaN with NG in below mentioned columns
column = ['location','department','salary_range','company_profile','description','requirements','employment_type','industry','function']
for i in column:
    job[i].fillna('NG',inplace=True)
```

```python
# Replace all NaN with 0 and make all data to 1 whiich is not null in below mentioned columns
column = ['benefits','required_experience','required_education']
for i in column:
    job[i][~job[i].isnull()] = 1
    job[i][job[i].isnull()] = 0
```

```python
X = job[features]
y = job[target]
```
